### Cum reprezentm cuvintele?

Prima problem pe care o vom trata este reprezentarea cuvintelor. Cum putem reprezenta un cuvnt ntr-un format care s fie uor de procesat de ctre un computer?

#### Reprezentarea cuvintelor folosind one-hot encoding

Reprezentarea cuvintelor folosind one-hot encoding presupune definirea unui vocabular de cuvinte i atribuirea fiecrui cuvnt unui index unic. Reprezentarea unui cuvnt este reprezentat prin un vector de lungimea vocabularului, care are valoarea **1 pe poziia cuvntului i 0 pe poziiile celorlalte cuvinte** (de unde i one-hot).

Probleme:
+ vocabularul poate fi foarte mare
+ reprezentarea este foarte sparse
+ adugarea unui nou cuvnt n vocabular necesit reprezentarea tuturor cuvintelor din vocabularul anterior i mrirea dimensiunii vectorilor de reprezentare
+ cuvintele au aceeai reprezentare, indiferent de context - nu se tine cont de semnificaia cuvntului
+ toate cuvintele sunt considerate egale - nu exist relaii ntre acestea

n exemplul de mai jos, vom demonstra cum ar arta one-hot encoding aplicat la nivel de caractere. Acelai principiu poate fi ulterior aplicat la nivel de cuvinte, acest exemplu fiind unul pur demonstrativ pentru a nelege magnitudinea reprezentrii rezultate n urma acestui tip de conversie.

```python
# definim inputul
print("Textul original:")
data = "Pisica mea s-a urcat pe scaun."
print(data)
print()

# rescriem input cu litera mica (pt a usura conversia la intreg)
data = data.lower()
print("Text dup lowercasing:")
print(data)
print()

# definim toate caracterele posibile pe care vrem sa le avem in vedere
alphabet = 'abcdefghijklmnopqrstuvwxyz !"#$%&\'()*+,-./0123456789:;<=>?@[\\]^_`{|}~'
```

```python
# definim un dicionar ce ajut la conversia valorilor ca ntregi
char_to_int = dict((c, i) for i, c in enumerate(alphabet))

# definim un dicionar ce ajut la conversia valorilor ca i caractere folosind alfabetul
int_to_char = dict((i, c) for i, c in enumerate(alphabet))
```

```python
import numpy as np

# input convertit ca ntregi
integer_encoded = np.array([char_to_int[char] for char in data])
print("Text convertit la ntregi:")
print(integer_encoded)
print()

# refacem input din ntregi
recovered_string = np.array([int_to_char[integer] for integer in integer_encoded])
print("Conversia napoi la caractere:")
print(recovered_string)
```

```python
from numpy import argmax

# one hot encoding
onehot_encoded = list()
for value in integer_encoded:
	letter = [0 for _ in range(len(alphabet))]
	letter[value] = 1
	onehot_encoded.append(letter)

print("One-hot encoded text:")
with np.printoptions(threshold=np.inf,linewidth=np.inf):
  print(np.array(onehot_encoded))
  print()

# refacem textul original
print("Text recuperat din one-hot encoding:")
inverted_text = ''
inverted_encoding = [int_to_char[argmax(onehot_encoded_char)] for onehot_encoded_char in onehot_encoded]
print(inverted_text.join(inverted_encoding))
```

#### Reprezentarea cuvintelor folosind proprieti (word feature representation)

Reprezentarea cuvintelor folosind proprieti (word feature representation) presupune definirea unui set de proprieti i atribuirea fiecrui cuvnt a unui vector de lungimea setului de proprieti. Fiecare vector care are `valori reale pentru fiecare dintre proprietile definite` implicnd faptul ca `numrul proprietilor va oferi dimensionalitatea vectorilor de reprezentare`. Aceast metod de reprezentare `face posibil reprezentarea relaiilor dintre cuvinte`. Relaiile dintre cuvinte pot fi reprezentate prin:
+ `distane ntre vectorii de reprezentare` - distana euclidian
+ `similitudini ntre vectorii de reprezentare` - cosinusul unghiului dintre vectorii de reprezentare (`cosine similarity`)

Avantaje:
+ reprezentarea mai eficient a cuvintelor - fiecare cuvnt are un vector de lungimea setului de proprieti
+ cuvintele similare au reprezentari vectoriale similare
+ putem reprezenta relaiile dintre cuvinte - distane ntre vectorii de reprezentare are sens
+ putem aduga cuvinte noi n vocabular fr a mri reprezentarea vectorial a cuvintelor existente


Probleme:
+ setul de proprieti poate fi foarte mare
+ definirea proprietilor este dificil (feature engineering)

#### Reprezentarea cuvintelor folosind word embeddings

Problema reprezentrii cuvintelor folosind proprieti (word feature representation) este definirea setului de proprieti. Pentru a defini un cuvnt pot fi necesare sute de proprieti incluznd proprieti neinterpretabile de ctre om.

Word embeddings sunt reprezentri dense ale cuvintelor, care au fost obinute prin antrenarea unor reele neurale pe un corpus mare de texte. Word embeddings sunt reprezentri dense.

Exemple de word embeddings antrenate pe corpusuri mari de texte:
+ [GloVe](https://nlp.stanford.edu/pubs/glove.pdf)
+ [Word2Vec](https://medium.com/@zafaralibagh6/a-simple-word2vec-tutorial-61e64e38a6a1)
+ [FastText](https://amitness.com/2020/06/fasttext-embeddings/)

Avantaje:
+ reprezentare eficient
+ surprind cuvintele similare
+ surprind relaiile dintre cuvinte
+ putem antrena word embeddings pentru diferite task-uri individuale sau putem antrena un model generalist pentru a rezolva mai multe task-uri simultan

Probleme:
+ word embeddings sunt sensibile la context (contextul n care au fost antrenate VS contextul n care vor fi folosite)
+ limbi mai slab/bine reprezentate (e.g. malga vs. englez)
+ domenii mai bine/slab reprezentate (e.g. tehnologie vs. dendrologie)


n urmatoarea celul vom folosi embedding-uri preantrenate folosind biblioteca gensim.


```python
import gensim
import gensim.downloader
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# reprezentari vectoriale preantrenate - word embeddings (ultimul numar reprezinta dimensiunea vectorului)
'''
 'fasttext-wiki-news-subwords-300',
 'conceptnet-numberbatch-17-06-300',
 'word2vec-ruscorpora-300',
 'word2vec-google-news-300',
 'glove-wiki-gigaword-50',
 'glove-wiki-gigaword-100',
 'glove-wiki-gigaword-200',
 'glove-wiki-gigaword-300',
 'glove-twitter-25',
 'glove-twitter-50',
 'glove-twitter-100',
 'glove-twitter-200',
 '__testing_word2vec-matrix-synopsis'
 '''
pretr_emb_model = gensim.downloader.load('glove-wiki-gigaword-50')

word_set = [
            "man", "woman", "girl", "king", "queen", "boy", "prince", "princess",
            "programming", "coding", "developer", "laptop", "server", "phone", "algorithm",
            "edema", "fracture", "acute", "contusion",
            "car", "truck", "boat"
            ]

# extragem reprezentarile vectoriale preantrenate
emb_set = np.array([pretr_emb_model[x] for x in word_set])


print("Dimensionalitatea reprezentarilor vectoriale:", emb_set.shape)

# aplicam PCA pentru a reduce dimensionalitatea
dim_red_model = PCA(n_components=2)
emb_set = dim_red_model.fit_transform(emb_set)

# plot
plt.figure(figsize=(12, 12))
plt.plot(emb_set[:, 0], emb_set[:, 1], 'o')
for i, word in enumerate(word_set):
    plt.annotate(word, (emb_set[i, 0], emb_set[i, 1]))
plt.title("GloVe word embeddings")
plt.show()

```

Text cleaning presupune:
+ normalizarea textului (lowercase) - pentru a evita ca cuvintele s fie considerate diferite datorit diferenei de case
+ eliminarea caracterelor speciale (punctuaie, simboluri, emoji-uri, etc) - pentru a evita ca cuvintele s fie considerate diferite datorit acestora
+ eliminarea numerelor
+ eliminarea sintagmelor ce nu sunt relevante analizei (i acestea sunt dependente de task) - link-uri, adrese de email, etc

```python
text_suport = "HEY amazon - my package never arrived https://www.amazon.com/order please fix asap! I have paid 1000$ for the product but it is not delivered yet. Get your things in order, Amazon. @amazonhelp"

print("=" * 50)
print("Textul initial:")
print("-" * 50)
print(text_suport)
print("=" * 50)

# normalizarea textului
text_suport = text_suport.lower()
print("Dupa normalizare:")
print("-" * 50)
print(text_suport)
print("=" * 50)

# eliminarea linkurilor
text_suport = re.sub(r'http\S+', '', text_suport)
print("Dupa eliminarea linkurilor:")
print("-" * 50)
print(text_suport)
print("=" * 50)

# fiind un tweet, eliminam @username
text_suport = re.sub(r'@\S+', '', text_suport)
print("Dupa eliminarea @username:")
print("-" * 50)
print(text_suport)
print("=" * 50)

# eliminarea caracterelor speciale si a numerelor (pastram doar textul)
text_suport = re.sub(r'[^a-zA-Z\s]', '', text_suport)
print("Dupa eliminarea caracterelor speciale si a numerelor :")
print("-" * 50)
print(text_suport)
print("-" * 50)

```

#### Tokenizarea textului (Tokenization)

Tokenizarea se refer la procesul de mprire a blocului de text ntr-o list de cuvinte sau token-uri. Putem mpri paragraful n propoziii, propoziia n cuvinte sau cuvntul n caractere. Aceasta impartire in token-uri faciliteaza interpretarea textului.

Tokenizarea este necesar pentru a putea aplica algoritmi de NLP asupra textului.



```python
# tokenizarea textului
text_suport = word_tokenize(text_suport)

print("Dupa tokenizare:")
print("-" * 50)
print(text_suport)
print("=" * 50)
```

#### Eliminarea stopword-urilor (Stopwords removal)

Stopword-urile sunt cuvinte care sunt filtrate atunci cnd se execut sarcini de NLP. Stopword-urile sunt cuvinte care apar din abunden i **nu ofer nicio informaie** modelului nostru. Stopword-urile care vor fi eliminate depind de sarcina care va fi executat.

Dezavantajul eliminrii stopword-urilor este c acest pas de preprocesare **poate afecta contextul** blocului de text.

```python
# stopword-urile standard din pachetul nltk pe engleza
stop_words
```

```python
print("Inainte de eliminarea stopword-urilor:")
print("-" * 50)
print(text_suport)
print("=" * 50)

# eliminarea stopword-urilor
tok_text_suport = [word for word in text_suport if word not in stop_words]

print("Dupa eliminarea stopword-urilor:")
print("-" * 50)
print(tok_text_suport)
print("=" * 50)
```

Se poate observa c `eliminarea stopword-urilor poate afecta contextul`. Stopword-ul "not" este eliminat i "it is not delivered" devine "it is delivered". Setul de stopword-uri poate fi customizat n funcie de task-ul ce trebuie rezolvat.

#### Stemming

Stemming este procesul de **reducere a cuvintelor derivate/cu sufix/cu prefix**, astfel nct acestea s se reduc la radcina lor/ cuvntul lor de baz. Stemming-ul elimin ultimele caractere dintr-un cuvnt, `ducnd adesea la semnificaii i ortografie incorecte.` Stemming-ul este `preferat pentru dataset-uri mari` cnd performana este mai important dect acurateea.


Pentru tehnicile de stemming cele mai comune opiuni sunt: Porter i Lancaster (nu intrm n detalii despre cum funcioneaz). Pentru exemplu vom folosi Porter.


```python
porterStemmer = PorterStemmer()

print("Inainte de stemming:")
print("-" * 50)
print(tok_text_suport)
print("=" * 50)

# stemming cu Porter
stem_text_suport = [porterStemmer.stem(word) for word in tok_text_suport]
print("Dupa stemming cu Porter:")
print("-" * 50)
print(stem_text_suport)
print("=" * 50)
```

#### Lemmatizarea (Lematization)

Lematizarea este o tehnic **contient de context** care reduce cuvintele la forma lor de baza i utilizeaz reguli diferite de normalizare pentru diferite pri ale vorbirii.

Lematizarea este de preferata fa de Stemming deoarece lematizarea **face analiza morfologic a cuvintelor (ia n considerare contextul morfologic) i pstreaz nelesul cuvntului (forma de baz a cuvntului are sens)**.

Efectuarea pailor de preprocesare rezult n generarea unui vocabular concis de o dimensionalitate mult mai mic dect cea a textului iniial.

```python
wordNetLematizer = WordNetLemmatizer()

print("Inainte de lemmatizare:")
print("-" * 50)
print(tok_text_suport)
print("=" * 50)

# lemmatizare
lem_text_suport = [wordNetLematizer.lemmatize(word) for word in tok_text_suport]
print("Dupa lemmatizare:")
print("-" * 50)
print(lem_text_suport)
print("=" * 50)
```

```python
# comparatie intre stemming si lemmatizare
print("Comparatie intre stemming si lemmatizare:")
print("-" * 50)
print("Stemming   :", stem_text_suport)
print("Lemmatizare:", lem_text_suport)
print("=" * 50)
```

#### Bag of Words (BoW)

Dup efectuarea pailor de preprocesare a textului, acum vom reprezenta textul sub forma unui vector de frecvene.

```python
from collections import Counter

bow = Counter(lem_text_suport)

# numararea cuvintelor
print("Reprezentarea folosind Bag-Of-Words:")
print("-" * 50)
print(bow)
print("=" * 50)
```

#### mbuntirea reprezentrii textului

Pn acum am folosit reprezentarea textului sub forma unui `vector de frecvene pentru cuvinte (Unigrame)`.

Aceast reprezentare este limitat deoarece nu ia n considerare contextul. Pentru a lua n considerare i contextul, vom folosi reprezentarea textului sub forma unui `vector de frecvene pentru cuvinte i perechi de cuvinte (bigrame)`. Bigramele sunt reprezentate de cuvinte/token-uri consecutive. `ncorpornd i contextul n reprezentarea textului, performana modelului ce va folosi aceast reprezentare va fi mai bun.`

Metoda poate fi extins i la `n-grame`.

```python
bow_bi = Counter(nltk.bigrams(lem_text_suport))

print("Reprezentarea folosind Bag-Of-Words")
print("-" * 50)
print(bow)
print("=" * 50)

print("Reprezentarea folosind Bag-Of-Words pentru bigrame:")
print("-" * 50)
print(bow_bi)
print("=" * 50)
```

#### Exerciiul 2 - Folosirea reprezentrilor generate pentru sentiment analysis (3p)

5. Testai modelul pe setul de date de testare i raportai acurateea obinut. (1p)

```python
# CounterVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import sklearn.metrics as metrics

cv = CountVectorizer(max_features=500, ngram_range=(1, 2))
data_cv = cv.fit_transform(data["tweet"].apply(lambda x: ' '.join(x)))

train_X, test_X, train_y, test_y = train_test_split(data_cv, data["label"], test_size=0.2, random_state=42)

cs = MultinomialNB()

# fit and transform
cs.fit(train_X, train_y)

# predict
pred_y = cs.predict(test_X)

# accuracy
print("Accuracy:", metrics.accuracy_score(test_y, pred_y))
print("Confusion matrix:")
print(metrics.confusion_matrix(test_y, pred_y))
```

## Arhitectura de tip Transformer

La nivel general, arhitectura unui Transformer arat astfel:

![transformer arch](https://machinelearningmastery.com/wp-content/uploads/2021/08/attention_research_1.png)



La modul genral, aceast arhitectur este destul de simpl, fiind compus din 2 mari componente:
+ straturi de tip feed-forward (asemntoare unei reele neurale simple)
+ blocuri de atenie (n acest caz este vorba de atenia la mai multe capete - multi-head attention)

Aceste componente sunt combinate ingenios alturi de cteva alte concepte cum ar fi legturi reziduale (skip connections) i desigur un mod inteligent de reprezentare a textului (unde dorim s cunoatem att poziia - positional embedding - ct i textul care este prezent la poziia respectiv - cu un input embedding ce rezult n urma tokenizrii)

Structura este replicat att la nivelul encoderului ct i la nivelul decoderului, cele 2 blocuri (reprezentate n figur n paralel) fiind aproape identice.

Un bun ghid ilustrativ pentru aceast arhitectur poate fi gsit [aici](https://jalammar.github.io/illustrated-transformer/). Tot aici vei gsi i o explicaie n detaliu a tuturor componentelor de figur. Aceasta servete n cadrul laboratorului cu scop ilustrativ, toate componentele acesteia alturi de matemtatica din spate putnd face obiectul unui curs de sine stttor.

nainte de a trece ctre seciunea practic i a nva cum putem utiliza astfel de modele, dorim s mai facem o scurt trecere n revist a celor mai populare 2 versiuni pentru arhitectura de tip Transformer.

### Modele de tip Transformer formate doar din decoder

O limitare important a modelelor de tip encoder-only este reprezentat de incapacitatea acestora de a genera un semnal de ieire coerent (fie c vorbim de generarea unei poveti, a unei picturi sau a unui cntect - n funcie de datele pe care a fost antrenat). Aparent aceast reprezentare este foarte potrivit pentru a comprima textul dar nu i pentru a l genera.

n acest sens, OpenAI, tot n 2018, lanseaz arhitectura de tip GPT (Generative Pre-trained Transformer). Dei aceast arhitectur a avut o evoluie mai lent, nefiind imediat evidente avantajele sale, aceasta este paradigma pe care am ajuns s o folosim n prezent la scar larg.

+ introdu tot acest text ntr-un model i f o trecere complet prin acest corpus pt a antrena modelul (folosind o metod de antrenare nesupervizat/semi-supervizat - Masked Language Modelling)
+ folosete acest model pentru a genera informaie, o unitate/token pe rnd

## Strategii de sampling

Mai sus, ne-am folosit de strategia de sampling oferit de metoda generate() din HuggingFace. Fr nici o modificare, la fiecare pas, ea ne va returna cel mai probabil token de la acel pas.

Tokenul cu "cea mai mare probabilitate" nu este ns mereu cel mai dezirabil. De exemplu, rspunurile obinute pot ajunge s fie foarte repetitive sau am putea constata c obinem mai mereu rspunsuri inventate / halucinaii ale modelului.

ntr-un anumit sens, strategia de sampling poate mbunti calitatea rspunului la modul general ns nu poate garanta c rspunsurile obinute vor fi mereu cele mai corecte i actuale. Astfel de caracteristici pot fi controlate doar controlnd datele care ajung s fie vzute de model (de exemplu n timpul antrenrii)

```python
#prima ncercare: nlocuirea strategiei de sampling
generation_config = GenerationConfig(
    do_sample=True,
)
question = input("Enter a question here: ")
prompt = f"""<|system|>
You are an assistant answering questions in the voice of a pirate.
</s>
<|user|>
{question}</s>
<|assistant|>"""
inputs = tokenizer(prompt, return_tensors="pt")
input_ids = inputs["input_ids"].cuda()
generation_output = model.generate(
    input_ids=input_ids,
    generation_config=generation_config,
    return_dict_in_generate=False,
    pad_token_id = tokenizer.eos_token_id,
    max_new_tokens=100,
)
answer = tokenizer.decode(generation_output[0])
print("Answer:", answer)
```

```python
#a doua ncercare: modificarea temperaturii
generation_config = GenerationConfig(
    do_sample=True,
    temperature = 0.8,
)
question = input("Enter a question here: ")
prompt = f"""<|system|>
You are an assistant answering questions in the voice of a pirate.
</s>
<|user|>
{question}</s>
<|assistant|>"""
inputs = tokenizer(prompt, return_tensors="pt")
input_ids = inputs["input_ids"].cuda()
generation_output = model.generate(
    input_ids=input_ids,
    generation_config=generation_config,
    return_dict_in_generate=False,
    pad_token_id = tokenizer.eos_token_id,
    max_new_tokens=100,
)
answer = tokenizer.decode(generation_output[0])
print("Answer:", answer)
```

## LLMs ca interfata pentru algoritmi de clasificare



```python
import matplotlib.pyplot as plt
question = "Is looking at the logits of answers a good way to extract the knowledge of large language models?"
prompt = f"""<|system|>
You are an assistant answering questions as truthfully as possible. Think step by step.
</s>
<|user|>
{question}</s>"""
inputs = tokenizer(prompt, return_tensors="pt")
input_ids = inputs["input_ids"].cuda()

res = model.forward(input_ids=input_ids,)
print("The shape of the inputs is:", input_ids.shape)
print("The shape of the logits is:", res.logits.shape)
# This result of the forward pass has three dimensions.
# * A batch dimension
# * A sequence dimension (the input had this many tokens)
# * A vocab dimension (the model can choose between this many tokens)
next_token_logits = res.logits[0, -1, :].cpu()
next_token_prob = softmax(next_token_logits.detach().numpy(), axis=-1)
plt.plot(range(len(next_token_prob)), next_token_prob)
plt.xlabel("Token index")
plt.ylabel("Probability")
plt.show()
```

```python
#@title Afisam tokens cei mai probabili
# Take the 20 most probable tokens which come after the question we asked.
highest_prob_tokens = np.argsort(next_token_prob, axis=0)[-1:-50:-1]
# Print those tokens
print([(tokenizer.decode(i), i) for i in highest_prob_tokens])
```
